step #1: Create a collection for video searching

```python
# Get API key from `https://trytldw.ai/account?tab=api`
API_KEY = '' # Insert API key here!!!
API_ENDPOINT = 'https://api.trytldw.ai/v1'
```

```python
import requests

headers = {
    'Authorization': f'Bearer {API_KEY}',
}
res = requests.post(f'{API_ENDPOINT}/collections', headers=headers, json={'name': 'Disaster Ariel Footage'})

print(res.json())
assert res.status_code == 200

collection_id = res.json()['collection_id']
# collection_id = '2f86d6f3-0699-4d5b-a3d3-71433b073cbc'
```

step #2: Adding videos to collection and building visual index on it

```python
# collection_id = '2f86d6f3-0699-4d5b-a3d3-71433b073cbc'

res = requests.post(f'{API_ENDPOINT}/media', headers=headers, json={
    'collection_id': collection_id,
    'media_list': [
        {
            'title': 'Destroyed Building',
            'external_id': 'video_0',
            'url': 'https://storage.googleapis.com/test-only-public/Aerial%20footage%20shows%20the%20destruction%20in%20Asheville%20and%20parts%20of%20western%20North%20Carolina%20%5BiLlebC7vuJw%5D.webm'
        },
        {
            'title': 'Explosion',
            'external_id': 'video_1',
            'url': 'https://storage.googleapis.com/test-only-public/Drone%20video%20shows%20deadly%20fireworks%20explosion%20in%20the%20Honolulu%20area%20%5BQbbXlMyq3Cs%5D.webm'
        },
        {
            'title': 'Wildfire Smoke',
            'external_id': 'video_2',
            'url': 'https://storage.googleapis.com/test-only-public/Watch%20from%20above%EF%BC%9A%20Aerial%20footage%20of%20natural%20disasters%20in%202017%20%5BmKWq9m3rkTU%5D.mp4'
        }
    ]
})

print(res.json()['media_list'])
media_ids = [media['id'] for media in res.json()['media_list']]
```

step #3: Wait until all indexing jobs are done

```python
# media_ids = ['f9506fb1-5803-4c12-8004-f5401e331bbb', '4e18b554-7772-4d9c-a5a7-275c2e314997', 'e4fb5388-8f11-4bbb-9f01-ace98b1f45e1']

status_list = []
for id in media_ids:
    status = requests.get(f'{API_ENDPOINT}/media/{id}', headers=headers).json()['status']
    print(f'media {id} status: {status}')
    status_list.append(status)

ready = all([status == 'COMPLETED' for status in status_list])
print(f'ready: {ready}')
```

step #4: Search in the collection

```python
res = requests.post(f'{API_ENDPOINT}/query/collection', headers=headers, json={
    'collection_id': collection_id,
    'query': 'building collapsed'
})

import json
print(json.dumps(res.json(), indent=2))

# Single segment
# {
#         "media_uuid": "90a8d329-1213-4c7d-a40d-bdf73d432c7e",
#         "segment_start_ms": 286000.0,
#         "segment_end_ms": 288000.0,
#         "score": 0.3966263234615326,
#         "description": null
#       },
```
